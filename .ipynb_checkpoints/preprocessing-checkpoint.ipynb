```python
def gen_sequence(Tokenizer):
    file = open('data/shakespeare.txt', 'r')
    s = file.read()
    sentences = s.split('\n')
    lines = []
    for sentence in sentences:
        temp = keras.preprocessing.text.text_to_word_sequence(sentence,
                                                       filters='!"#$%&()*+,./:;<=>?@[\\]^_`{|}~\t\n0123456789',
                                                       lower=True,
                                                       split=' ')
        if len(temp) > 1:
            lines.append(temp)
        # print(lines)

    token_list = []
    
    for line in lines:
        fin_list = []
        temp_list = Tokenizer.texts_to_sequences(line)
        
        for num in temp_list:
            fin_list.append(num[0] - 1)
            
        token_list.append(fin_list)
        
    print(token_list)
    return token_list
```
